```python
from db_entities_and_session import db
from AcmDlArticleParser import AcmDlArticleParser
from AcmDlListOfArticlesParser import AcmDlListOfArticlesParser
from AcmDlListOfProceedingsParser import AcmDlListOfProceedingsParser
from pony.orm import *
import misc
import re
```

```python
@db_session
def commit_in_db(db_session, articles_data):
    for article in articles_data:
        venue_entity = db_session.Venue.get(name=article['venue']['name'])
        if not venue_entity:
            venue_entity = db_session.Venue(name=article['venue']['name'])
        db_session.commit()
        venue_id = db_session.Venue.get(name=article['venue']['name']).venue_id
        
        db_session.Article(article_id=article['article_id'],
                           title=article['title'],
                           abstract=re.sub('[\'\']', '', repr(article['abstract'])),
                           year=article['year'],
                           venue_id=venue_id)
        
        authors = article['authors']
        for author in authors:
            author_entity = db_session.Author.get(name=author['name'])
            if not author_entity:
                author_entity = db_session.Author(name=author['name'], url=author['url'])
            db_session.commit()
            author_id = db_session.Author.get(name=author['name']).author_id
            
            db_session.AuthorArticle(author_id=author_id, article_id=article['article_id'])
        
        for citation in article['cited_by']:
            citation_entity = db_session.Citation(article_id=article['article_id'], cited_by=citation)
        
        # for cited_by in article['cited_by']:
        #    db_session.Citation(article_id=article['article_id'], cited_by=cited_by)
```
