```python
from db_entities_and_session import db
from AcmDlArticleParser import AcmDlArticleParser
from AcmDlListOfArticlesParser import AcmDlListOfArticlesParser
from AcmDlListOfProceedingsParser import AcmDlListOfProceedingsParser
from pony.orm import *
import misc
import re
```

```python
@db_session
def commit_in_db(db_session, article_data):
    article_entity = db_session.Article.get(article_id=article_data['article_id'])
    
    if not article_entity:
        venue_entity = db_session.Venue.get(name=article_data['venue']['name'])
        if not venue_entity:
            venue_entity = db_session.Venue(name=article_data['venue']['name'])
        db_session.commit()
        venue_id = db_session.Venue.get(name=article_data['venue']['name']).venue_id

        db_session.Article(article_id=article_data['article_id'],
                           title=article_data['title'],
                           abstract=re.sub('[\'\']', '', repr(article_data['abstract'])),
                           year=article_data['year'],
                           venue_id=venue_id)

        authors = article_data['authors']
        for author in authors:
            author_entity = db_session.Author.get(name=author['name'])
            if not author_entity:
                author_entity = db_session.Author(name=author['name'], url=author['url'])
            db_session.commit()
            author_id = db_session.Author.get(name=author['name']).author_id

            db_session.AuthorArticle(author_id=author_id, article_id=article_data['article_id'])

        for citation in article_data['cited_by']:
            citation_entity = db_session.Citation(article_id=article_data['article_id'], cited_by=citation)
        
        db_session.commit()
```

```python
event_parser = AcmDlListOfProceedingsParser()
proceeding_ids = event_parser.parse(misc.event_id, misc.client_id)
```

```python
number_of_articles_in_event = 0

start_proceeding = 0
end_proceeding = 50

article_ids_nested = []

for proceeding_id in proceeding_ids[start_proceeding:end_proceeding]:
    print('Progress: {0}\{1}'.format(proceeding_ids.index(proceeding_id)+1, len(proceeding_ids)))
    
    proceeding_parser = AcmDlListOfArticlesParser()
    article_ids_in_proceeding = proceeding_parser.parse(proceeding_id)
    
    article_ids_nested.append(article_ids_in_proceeding)
    number_of_articles_in_event += len(article_ids_in_proceeding)
```

```python
number_of_articles_in_event
```

```python
article_ids = []

for ls in article_ids_nested:
    for article_id in ls:
        article_ids.append(article_id)
        
len(article_ids)
```

```python
# articles_data = []
success_counter = 0

for article_id in article_ids:
    print('Progress: {0}\{1}'.format(article_ids.index(article_id)+1, number_of_articles_in_event))
    try:
        article_parser = AcmDlArticleParser()
        article_data = article_parser.parse(article_id)
        articles_data.append(article_data)
        success_counter += 1
        print('\n\n')
    except Exception as e:
        print(str(e))
        print('Parsing article with id: {0} is not successful'.format(article_id))
print('{0} from {1} articles were parsed successfully'.format(success_counter, number_of_articles_in_event))
```
