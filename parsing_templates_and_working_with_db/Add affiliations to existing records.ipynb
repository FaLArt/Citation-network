```python
from db_dumps_and_entities.db_entities_and_session import db
from pony.orm import db_session

from parsing_scripts.AcmDlArticleParser import AcmDlArticleParser
```

```python
@db_session
def commit_affiliation_in_db(article_data):
    article_id = article_data['article_id']
    
    for author_and_affiliation in article_data['authors_and_affiliations']:
        
        author_entity = db.Author.get(name=author_and_affiliation['name'])
        if not author_entity:
            author_entity = db.Author(name=author_and_affiliation['name'], url=author_and_affiliation['url'])
            db.commit()
        author_id = author_entity.author_id
        
        affiliation_entity = db.Affiliation.get(url=author_and_affiliation['affiliation']['url'])
        if not affiliation_entity:
            affiliation_entity = db.Affiliation(name=author_and_affiliation['affiliation']['name'],
                                               url=author_and_affiliation['affiliation']['url'])
            db.commit()
        affiliation_id = affiliation_entity.affiliation_id
        
        aa_entity = db.AuthorArticle.get(author_id=author_id, article_id=article_id)
        if not aa_entity:
            aa_entity = db.AuthorArticle(author_id=author_id, article_id=article_id, affiliation_id=affiliation_id)
            db.commit()
        else:
            aa_entity.affiliation_id = affiliation_id
            
    print('Article with ID {0} was editted successfully ^_^'.format(article_id))
```

```python
total = len(db.Article.select()[360:len(db.Article.select())])
total_counter = 0
success_counter = 0

for article in db.Article.select()[360:len(db.Article.select())]:
    print('Progress {0}\{1}'.format(total_counter + 1, total))
    try:
        parser = AcmDlArticleParser()

        article_id = article.article_id

        article_data = parser.parse(article_id=article_id)

        commit_affiliation_in_db(article_data)
        
        success_counter += 1
    except Exception as e:
        print(e)
    total_counter += 1
    
    print('\n')

print('{0}\{1}'.format(success_counter, total_counter))
```
